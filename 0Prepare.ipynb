```python
# 使用numpy + matplotlib 画一个正弦曲线
import numpy as np
x = np.linspace(0,2*np.math.pi,100)
y = [np.math.sin(i) for i in x]

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('default')
plt.plot(x,y)
```

```python
#修改matplotlib的样式
plt.style.use('fivethirtyeight')
# Set an aspect ratio， 比例，大小
width, height = plt.figaspect(1.2)
print width,height
fig = plt.figure(figsize=(width,height), dpi=400)
plt.plot(x,y)
```
