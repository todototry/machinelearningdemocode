```python
from sklearn.datasets import load_digits
digits = load_digits()
print("images shape: %s" % str(digits.images.shape))
print("targets shape: %s" % str(digits.target.shape))
print digits.viewkeys()
print "data:", digits.data
#print "images:", digits.images

%matplotlib inline
import matplotlib.pyplot as plt
plt.matshow(digits.images[0], cmap=plt.cm.Greys);
plt.matshow(digits.data[0].reshape(8,8), cmap=plt.cm.Greys);
```
